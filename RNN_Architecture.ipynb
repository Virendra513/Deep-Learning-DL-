 Recurrent Neural Networks (RNNs) are a type of artificial neural network designed for sequential data processing. They have an internal memory mechanism that allows them to maintain a hidden state, enabling them to capture information about previous inputs in the sequence. RNNs are widely used in various applications involving sequential or time-series data.



RNN.png

```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
# Basic RNN model
model= Sequential()

model.add(SimpleRNN(3,input_shape=(3,5)))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python
# First layer weights
print(model.get_weights()[0].shape)
model.get_weights()[0]
```

```python
#  Feedback layer weights
print(model.get_weights()[1].shape)
model.get_weights()[1]
```

```python
# Hidden layer Biases
print(model.get_weights()[2].shape)
model.get_weights()[2]
```

```python
# Weights btw hidden layer and output layer
print(model.get_weights()[3].shape)
model.get_weights()[3]
```

```python
# Bias of output layer
print(model.get_weights()[4].shape)
model.get_weights()[4]
```

Challenges and Variations:
Vanishing and Exploding Gradients: RNNs may suffer from the vanishing or exploding gradient problem, limiting their ability to capture long-term dependencies. Architectures like LSTMs and GRUs were introduced to address this issue.*italicized text*
